There are three ways of creating plot

```python
# First method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
```

Workflow of Matplotlib

```python
# Step-1 --- Import necessary libraries

import matplotlib.pyplot as plt

# Step-2 --- Get or collect data
x = [1,2,3,4]
y = [11,12,13,14]

# Step-3 --- Set plot
fig,ax = plt.subplots(figsize = (10,10))

# Step-4 --- plot it
ax.plot(x,y)

# Step-5 --- Customize it
ax.set(title = "Sample simple plot" , xlabel ="x-axis", ylabel="y-axis")

# Step-6 --- Save it
fig.savefig("one-image.png")
```

## Plotting with O-O method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             figsize=(10,5))
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(10,6))

scatter1 = ax.scatter(over_50["age"],over_50["chol"],c=over_50["target"])

ax.set(title="Heart Disease and Cholostrol Level", ylabel = "Cholostrol", xlabel="Age")

ax.legend(*scatter1.legend_elements(),title="Target")

ax.axhline(over_50["chol"].mean(),linestyle="--")
```

```python
# Plotting two subplots

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex =True)

scatter1 = ax0.scatter(over_50["age"],over_50["chol"],c=over_50["target"])

ax0.legend(*scatter1.legend_elements(),title="Target")

ax0.set(title="Heart Disease and Cholesterol", ylabel="Cholesterol")

ax0.axhline(over_50["chol"].mean(),linestyle="--")



scatter1 = ax1.scatter(over_50["age"],over_50["thalach"],c=over_50["target"])

ax1.legend(*scatter1.legend_elements(),title="Target")

ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max HeartBeat")

ax1.axhline(over_50["thalach"].mean(),linestyle="--")

fig.suptitle("Heart Analysis", fontsize=16, fontweight="bold")
```

```python
# Axis 0, 1 (row 0, column 0)
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 10))
scatter = ax0.scatter(over_50["age"],
                      over_50["chol"],
                      c=over_50["target"],
                      cmap='winter')
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Set the x axis
ax0.set_xlim([50, 80])

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(),
            color='r',
            linestyle='--',
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"],
                      over_50["thalach"],
                      c=over_50["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Set the y axis
ax1.set_ylim([60, 200])

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(),
            color='r',
            linestyle='--',
            label="Average");
ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
