# Getting Data Ready

```python
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease
```

## Spliting the datatypes

```python
# Create X(feature matrix)
X = heart_disease.drop("target", axis=1)

#Create y(labels)
Y = heart_disease["target"]
```

```python
X
```

```python
Y
```

# Choose the right model and hyperparameter

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()

model1.get_params()
```

# 3 Fit the model to the training data

```python
from sklearn.model_selection import train_test_split

X_train,X_text,Y_train,Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
model1.fit(X_train,Y_train)
```

```python
X_train
```

# Evaluting the model

```python
y_pred = model1.predict(X_text)
y_pred
```

```python
Y_test
```

```python
model1.score(X_train,Y_train)
```

```python
model1.score(X_text,Y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
print(classification_report(Y_test,y_pred))
```

```python
confusion_matrix(Y_test,y_pred)
```

```python
accuracy_score(Y_test,y_pred)
```

# Improve a model


```python
import numpy as np
np.random.seed(42)
for i in range(10,100,10):
  print(f"Trying model with {i} estimators...")
  model1 = RandomForestClassifier(n_estimators=i).fit(X_train, Y_train)
  print(f"Model accuracy on test set: {model1.score(X_text, Y_test) * 100:.2f}%")
  print("")
```

# Save model

```python
import pickle
```

```python
pickle.dump(model1, open("random_forest.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest.pkl", "rb"))
loaded_model.score(X_text, Y_test)
```
