```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()

housing
```

## Creating DataSet

```python
df = pd.DataFrame(housing['data'], columns=housing["feature_names"])
df.head()
```

```python
df["target"] =  housing['target']

df.head()
```

## Spliting DataSet for training

```python
x = df.drop("target",axis=1)
x.head()
```

```python
y = df["target"]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)

x_train.head()
```

```python
y_train.head()
```

## Training the model

```python
from sklearn.ensemble import RandomForestRegressor

model1 = RandomForestRegressor()
```

## Evaluation of Model

r2_squared

```python
from sklearn.model_selection import cross_val_score

cv_r2_squared = cross_val_score(model1, x, y, scoring=None)

cv_r2_squared
```

# Scoring Regression

```python
print(f"The r2 squared is: {np.mean(cv_r2_squared)}")
```

neg_mean_absolute_error

```python
cv_nmae = cross_val_score(model1,x,y,scoring="neg_mean_absolute_error")

cv_nmae
```

```python
print(f"The negative mean absolue error is: {np.mean(cv_nmae)}")
```

neg_mean_squared_error

```python
cv_nmse = cross_val_score(model1,x,y,scoring="neg_mean_squared_error")

cv_nmse
```

```python
print(f"The negative mean square error is: {np.mean(cv_nmse)}")
```
